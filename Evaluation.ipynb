```python
import numpy as np
import seaborn as sns

sns.set(style='whitegrid',font_scale=1.2)
plt.figure(figsize=(10,10))


labels = ['Baseline', 'AINN', 'AI']
close1 = [data['Baseline_Close_1'].median(), data['AINN_Close_1'].median(), data['AI_Close_1'].median()]
close2 = [data['Baseline_Close_2'].median(), data['AINN_Close_2'].median(), data['AI_Close_2'].median()]


x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots(figsize=(10,7))
rects1 = ax.bar(x - width/2, close1, width, label='Close1', color='#C7362C', alpha=1)
rects2 = ax.bar(x + width/2, close2, width, label='Close2', color='#F0A8A8', alpha=1)

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Median')
ax.set_title('')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()
```
